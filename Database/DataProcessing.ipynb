```python
def data_process(self):
    complete_data  = pd.read_excel(os.path.join(chemfolder,"xlsexport.xlsx"))
    data = complete_data.drop("Unnamed: 0", axis=1)
    keys = data.keys()
    data_dict = data.to_dict()
    id = data.get(["SampleID"])
    idx = list(data.index)
        
    sorted_list = sorted(data_dict['SampleID'], key = lambda x:(data_dict['SampleID'][x])) #sorting the data according to SampleID
    final_dict = {}
    for key in data.keys():
        final_dict[key] = []
    for idx,count in enumerate(sorted_list):
        if idx != len(sorted_list)-1:
            for key in final_dict.keys():
                final_dict[key].append(data[key][sorted_list[idx]])
        
    data_new = pd.DataFrame.from_dict(final_dict)
        
    id_new = data_new.get(["SampleID"])
    idx_new = list(data_new.index)
        
    #creating a list of indexes for similar SampleID's
    start_list = [] 
    for id_new,sample in enumerate(data_new['SampleID']):
        if(id_new!=0 and id_new!=len(idx_new)-1):
            if data_new['SampleID'][id_new]==data_new['SampleID'][id_new-1]:
                pass
            else:
                start_list.append(id_new)
        else:
            start_list.append(id_new)
        
    #creating a final dictionary by taking mean of similar SampleID values
    final_dict_new = {}
    for key in data_new.keys():
        final_dict_new[key] = []
    unchanged_keys = ['AnalysisProgramDescription_','SampleID', 'Bezeichnung']
    for idx_new,count in enumerate(start_list):
        if idx_new != len(start_list)-1:
            for key in final_dict_new.keys():
                if key not in unchanged_keys:
                    final_dict_new[key].append(data_new[key][start_list[idx_new]:start_list[idx_new+1]].mean())
                else:
                    final_dict_new[key].append(data_new[key][start_list[idx_new]])
      
    df = pd.DataFrame.from_dict(final_dict_new)
        
    return df
```
